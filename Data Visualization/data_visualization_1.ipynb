```python
# Bar Graph
```

Syntax:
1. (vertical bar graph) ie, bars are vertically oriented. plt.bar(x, height, [width])
* 'x' : **list of** x-co-ords where the bats are CENTRED
* 'height' : **list of** heights of each bat
* 'width' : width of each bar. Default = 0.8

2. (horizontal bar graph) ie, bars are horizontally oriented. plt.barh(x, height, [width])

```python
plt.bar(['A', 'B', 'C', 'D'], [20, 40, 10, 5], width = 0.4) # Default width = 0.8
```

```python
plt.barh(['A', 'B', 'C', 'D'], [20, 40, 10, 5], height = 0.4)
```

```python
course_list = ['CSE', 'Mech', 'EXTC', 'Aero', 'Metal'] # x-axis labels
barIndex_alongXaxis = np.arange(len(course_list))

girlsCount_barHeight = [100, 40, 60, 240, 300]
boysCount_barHeight = [200, 70, 30, 120, 60]

width_per_bar = 0.4

bar1 = barIndex_alongXaxis
bar2 = [(i + width_per_bar) for i in bar1]

# GIRLS: plt.bar(x, height, [width])
# BOYS: plt.bar(x, height, [width])

plt.bar(bar1, girlsCount_barHeight, width_per_bar, label = 'Girls')
plt.bar(bar2, boysCount_barHeight, width_per_bar, label = 'Boys')
plt.legend()

plt.xticks((bar1 + (width_per_bar/2)), course_list)

plt.title('Ratio of Gender in each Engineering Branch')
plt.xlabel('Branches')
plt.ylabel('Count', rotation = 0, labelpad = 30)
```
